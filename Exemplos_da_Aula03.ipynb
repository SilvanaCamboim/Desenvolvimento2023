```python
import json

json_string = """{
  "name": "John Doe",
  "age": 30,
  "isStudent": true,
  "courses": ["math", "history", "chemistry"],
 "address": {"street": "123 Main St", "city": "New York", "state": "NY", "postalCode": "10001"
 }}"""

try:
    json_object = json.loads(json_string)

    print(json_object["name"])
    print(json_object["age"])
    print(json_object["isStudent"])
    print(json_object["courses"][-1])
    print(json_object["address"])


except json.JSONDecodeError as e:
    print(f"Error while parsing JSON: {e}")
```

Exemplo 2 - Básico **GEOJSON**

```python
import json

geojson_string = '''
{
  "type": "Feature",
  "geometry": {
    "type": "Point",
    "coordinates": [ -49.29459646136154, -25.423575592567218]
  },
  "properties": {
    "nome": "Torre Panorâmica",
    "endereco": "Rua Professor Lycio Grein Castro Vellozo, 191 - Mercês, Curitiba - PR, 80710-6501"
  }
}
'''

geojson_object = json.loads(geojson_string)

print(geojson_object["type"])
print(geojson_object["geometry"]["type"])
print(geojson_object["properties"]["nome"])
```

Exemplo 3 - GEOJSON no Folium

```python
import folium

long = geojson_object["geometry"]["coordinates"][0]
lat = geojson_object["geometry"]["coordinates"][1]

m = folium.Map (location= [lat, long],zoom_start=14)

gjson_layer = folium.GeoJson(geojson_object).add_to(m)

m
```

```python
import requests

# Define the API endpoint and the bounding box for Curitiba
url = "https://overpass-api.de/api/interpreter"
bbox = "-49.4639,-25.5835,-49.1387,-25.3021"

# Define the query string
query = f'[out:json][timeout:500];(node["amenity"="school"]({bbox});way["amenity"="school"]({bbox});relation["amenity"="school"]({bbox}););out body;>;out skel qt;'

# Send the request to the Overpass API and get the GeoJSON response
response = requests.post(url, data=query)
geojson = response.json()

print (geojson)
```
